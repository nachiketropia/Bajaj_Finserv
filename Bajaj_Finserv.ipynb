```python
patient_details = [entry["patientDetails"] for entry in data]
missing_firstName = 0
missing_lastName = 0
missing_DOB = 0

for patient in patient_details:
    if not patient.get("firstName"):
        missing_firstName += 1
    if not patient.get("lastName"):
        missing_lastName += 1
    if not patient.get("birthDate"):
        missing_DOB += 1

total_entries = len(patient_details)
missing_firstName_percentage = (missing_firstName / total_entries) * 100
missing_lastName_percentage = (missing_lastName / total_entries) * 100
missing_DOB_percentage = (missing_DOB / total_entries) * 100
missing_firstName_percentage_rounded = round(missing_firstName_percentage, 2)
missing_lastName_percentage_rounded = round(missing_lastName_percentage, 2)
missing_DOB_percentage_rounded = round(missing_DOB_percentage, 2)

missing_firstName_percentage_rounded, missing_lastName_percentage_rounded, missing_DOB_percentage_rounded
```
